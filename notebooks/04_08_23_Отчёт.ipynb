## Плагин

Поддержка call-шага осуществляется через использование плагина, добавляющего класс PredictJanusGraphFactory (наследник JanusGraphFactory). В этом классе переопределён метод "open" так, чтобы возвращать PredictJanusGraph (наследник StandardJanusGraph). В этом классе переопределён метод "getServiceRegistry" так, чтобы регистрировать PredictService. Всё, что делает PredictService - это передача статических параметров (пока только статических) вызова в запрос к endpoint-сервису, работающему на каком-то отдельном сервере, и возвращение полученного предсказания. Т.е. сервис, по сути, является прослойкой между гремлин-сервером и сторонним сервисом, хранящим всю логику.
